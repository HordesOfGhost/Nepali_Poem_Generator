# Import Libraries

```python
import ast
import matplotlib.pyplot as plt

import pickle
```

# LSTM Loss Metric

```python
metric_list = []
with open("lstm_train.txt","r") as f:
    metric_list.extend(f.readlines())

literal_list = []
for epoch in metric_list:
    epoch = epoch.replace(",\n","")
    epoch = eval(epoch)
    literal_list.append(epoch)

# Extracting values for loss and accuracy
epochs = [entry['Epoch'] for entry in literal_list]
loss_values = [entry['loss'] for entry in literal_list]
accuracy_values = [entry['accuracy'] for entry in literal_list]

# Plotting the loss
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(epochs, loss_values, label='Loss')
plt.title('Loss vs Epoch')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

# Plotting the accuracy
plt.subplot(1, 2, 2)
plt.plot(epochs, accuracy_values, label='Accuracy', color='orange')
plt.title('Accuracy vs Epoch')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

# Display the plots
plt.tight_layout()
plt.show()
```

# Transformer Metrices 

```python
file = 'loss_per_epoch_char.pkl'
with open(file, 'rb') as file:
    loss_value_char = pickle.load(file)

file = 'loss_per_epoch_word.pkl'
with open(file, 'rb') as file:
    loss_value_word = pickle.load(file)
```

```python
epochs_char = [epoch for epoch in range(0,len(loss_value_char))]
epochs_word = [epoch for epoch in range(0,len(loss_value_word))]
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(epochs_char, loss_value_char, label='Loss')
plt.title('Transformers Tokenized with Character')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

# Plotting the accuracy
plt.subplot(1, 2, 2)
plt.plot(epochs_word, loss_value_word, label='Loss')
plt.title('Transformers Tokenized with Word')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

# Display the plots
plt.tight_layout()
plt.show()
```
